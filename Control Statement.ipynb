# if statement 
- This statement is used to execute one or more statements depending on whether condition is true or not


     

```python
# to understand if 
num=1
if num==1:
    print('one')
```

# if else statement 
- The if else statement executes a group of statements when a condition is True otherwise it will execute another group of statements 
    

```python
# A Python program to test whether a number is negative or positive number
num=int(input("Enter the number: "))

if num>0:
    print("The numbers is positive")
else:
    print("The number is negative")
```

```python
# A Python program to test whether a given number is in between 1 and 10
num=int(input("Enter the number: "))

if num>0 and num<10:
    print("You entered",num,"which is between 1 and 10")
else:
    print("You entered",num,"Which is below 1 or above 10")
```

# if elif statement 
- sometimes, the programmer has to test multiple conditions and execute statements depending on those conditions. if elif else statement is useful in such situation 


```python
# A Python program to know if a given number is zero, positive or negative
num=int(input("Enter the number: "))
if num==0:
    print(num,"is zero")
elif num>0:
    print(num,"is positive")
else:
    print(num,"is negative")
```

```python
# A python program to print the grade depends upon the marks taken
#75% distinction
#60% first class
#50% second class
#35% third class
#below 35% failed 

marks=int(input("Enter the marks: "))

if marks>=75:
    print("distinction")
elif marks>=60 and marks<=74:
    print("First class")
elif marks>=50 and marks<=59:
    print("Second class")
elif marks>=35 and marks<=49:
    print("Third class")
else:
    print("failed")
```

# While loop
- A statements(if, if else) is executed only once from top to bottom
- Loop is useful to execute repeatedly
- The while loop is useful to execute a group of statements several times repeatedly depending on whether a condition is True or False


```python
# A Python program to display numbers from 1 to 10 using while loop
x=1
while x<=10:
    print(x)
    x+=1
print("end")
```

```python
# A Python program to display even numbers between m and n
m,n=[int(i) for i in input("Enter the minimum and maximum range").split(',')]
x=m
if x%2!=0:
    x=x+1
while x>=m and x<=n:
    print(x)
    x=x+2
```

# for Loop
- The for loop is udeful to iterate over the elements of sequence.
- It means the for loop can be used to execute a group of statements repeatedly depending on the number of element in the sequence 


```python
# A Python program to display charecters of a string using for loop
str="sadashiv"
for ch in str:
    print(ch)
```

```python
# A Python program to display odd numbers from 1 to 10 using range() function
for i in range(1,10,2):
    print(i)
    
```

```python
# A Python program to display and find the sum of a list of numbers using for loop
list=[10,20,30,40,50]
sum=0

for i in list:
    print(i)
    sum=sum+i
print("The sum of elements pressent in the list: ",sum)
```

# Nested loop
- It is possible to write one loop inside another loop
- for example we can write a for loop inside while loop or for loop inside another for loop, such loops are called 'nested loops'

```python
# A python program that display star in right angle triangular from using nested loops
for i in range(1,11):
    for j in range(1,i+1):
        print("*",end='')
    print()
```

# else suit
- In python it is possible to use 'else' statement along with for loop or while loop 
- The statement written after 'else' are called suits
- The else suit is always executed irrespective of the statements in the loop are executed or not


```python
for i in range(5):
    print("Yes")
else:
    print("No")
```

```python
# A Python program to search for an element in the list of elements
lst=[1,2,3,4,5]
search=int(input("Enter Element to search: "))
for element in lst:
    if search==element:
        print("Element found in list")
        break
else:
    print("Element not found in group")
```

# The break statement
- The break statement can be used inside a for loop or while loop to come out of the loop
- When 'break' is executed the python interpreter jumps out of the loop to process the next statement in the program


```python
# A Python program to display numbers from 10 to 6 and break the loop when the number about to display is 5
x=10
while x>=1:
    print('x= ',x)
    x=x-1
    if x==5:
        break
print("Out of loop")
```

# Continue Statement
- It returns the control to the beginning of the while loop.. 
- The continue statement rejects all the remaining statements in the current iteration of the loop and moves the control back to the top of the loop.
- The continue statement can be used in both while and for loops.


```python
str="Python"

for i in str:
    if i=='h':
        continue
    print(i)
```

# The assert statement
- The assert statement is useful to check if a particular condition is fulfilled or not. 
- The syntax is as follows **assert expression, message**
- In the above syntax, the 'message' is not compulsory


```python
# A program to handle the assertion error exception that is given by assert statement
x=int(input("Enter a number greater than 0: "))

assert x>0,"Wrong input entered"
print("You entered: ",x)
```

# The return statement
- return statement is used in function program to return output to the calling program

```python
# A python program to write a function that returns the result of sum of two numbers
def sum(a, b):
    c=a+b
    return c
#call the function
d=sum(2, 3)
print("The result is: ",d)
```
